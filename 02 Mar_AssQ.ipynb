# Problem 2:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
plt.figure(figsize = (5,4))
plt.scatter(x,y)
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show()
```

# Problem 3:

The "subplot()" funtion in matplotlib is used to display multiple plots or visualizations simultaneously in a single figure, arranged in rows and columns.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (6, 4))
axes[0,0].plot(x,y1)
axes[0,0].set_title('Line 1')
axes[0,1].plot(x,y2)
axes[0,1].set_title('Line 2')
axes[1,0].plot(x,y3)
axes[1,0].set_title('Line 3')
axes[1,1].plot(x,y4)
axes[1,1].set_title('Line 4')
plt.tight_layout()
plt.show()
```

# Problem 4:


A bar plot is a type of graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.

Bar plots are used to visualize and compare different categories of data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize = (4,3))
plt.bar(company, profit)
plt.show()
```

```python
plt.figure(figsize = (4,3))
plt.barh(company, profit)
plt.show()
```

# Problem 5:

A box plot is a type of statistical graph that shows the distribution of data. It is a quick and easy way to see the median, quartiles, and outliers of a data set.

Box plots are commonly used to analyze and compare data distributions across different categories or groups.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2], labels = ['box1', 'box2'])
plt.title("Box plot")
plt.show()
```
